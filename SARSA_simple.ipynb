# reward mapping

```python
def reward_mapping(next_state, reward, done, *args, **kwargs):
    # print("from, ns", reward, next_state)
    if done and reward == 1:
        reward = 200
    elif done:
        reward = 0
    else:
        reward = next_state
    # print("to", reward)
    return reward
```

```python
env = FrozenLakeEnvJP(is_slippery=True)
# strategy = QLearning(env, exploration_rate=0.5, learning_rate=0.7)
strategy = SARSA(env, exploration_rate=0.5, learning_rate=0.7)
play = Play(env, strategy, reward_mapping=reward_mapping)
```

```python
total_rewards = play.run(num_episodes=10000, is_render=False, verbose=False)
```

```python
print(strategy.exploration_rate)
strategy.show_state_action_matrix()

fig, ax = plt.subplots(figsize=(6, 6))  # Create a single plot
env.render_state_action_matrix(ax, strategy.state_action_matrix, grid_size=4)  # Use the function to plot
plt.show()  # Display the plot
```

```python
play.showcase()
```

```python
play.run(num_episodes=10)
```
