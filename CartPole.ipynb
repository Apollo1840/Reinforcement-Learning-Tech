```python
import gym
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time
```

```python
# Create the environment with render_mode='rgb_array' for compatibility
env = gym.make('CartPole-v1', render_mode='rgb_array')

# Function to render the environment and update the matplotlib figure
def render_env(env, step_number):
    img = env.render()  # Render and get the current frame as an image
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Step: {step_number}")  # Add the step number as a title to the plot
    display(plt.gcf())
    clear_output(wait=True)

# Reset the environment to the initial state
state = env.reset()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(6, 4))

# Run the environment for a set number of steps
for step in range(200):  # Run for 200 timesteps
    render_env(env, step)  # Render the environment
    
    # action = policy(state, ...)
    action = env.action_space.sample()  # Take a random action
    state, reward, done, truncated, info = env.step(action)  # Apply the action to the environment
    
    if done or truncated:
        # Reset the environment when the episode ends
        state = env.reset()
    
    time.sleep(0.01)  # Add a small delay to control the speed of the animation

# Close the environment when finished
env.close()
```
